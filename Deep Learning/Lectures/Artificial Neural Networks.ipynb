### Scaling Cols

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```

```python
df.sample(5)
```

## Input & Output Variables

```python
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
X.head()
```

```python
y.head()
```

## Training & Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

## Neural Network Model

```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(26, activation='relu', input_shape=(26,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
test_pred = model.predict(X_test)
```

```python
test_pred
```

```python
y_test
```

```python
test_pred_data = []
for data in test_pred:
    if data > 0.5:
        test_pred_data.append(1)
    else:
        test_pred_data.append(0)
```

```python
test_pred_data
```

### Model Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test, test_pred_data))
```

```python
print(confusion_matrix(y_test, test_pred_data))
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=test_pred_data)
```

```python
sns.heatmap(cm, annot=True, fmt='d')
```
