### Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading DataSet from Seaborn

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.lineplot(x = 'size', y='total_bill', data=df)
sns.set_style("darkgrid")
```

### Set Style

```python
sns.lineplot(x = 'size', y='total_bill', data=df)
sns.set_style("dark")
sns.despine()
```

```python
sns.lineplot(x = 'size', y='total_bill', data=df)
sns.set_style("white")
```

```python
sns.lineplot(x = 'size', y='total_bill', data=df)
sns.set_style("whitegrid")
```

```python
sns.lineplot(x = 'size', y='total_bill', data=df)
sns.set_style("ticks")
```

### Set Context

```python
sns.lineplot(x = 'size', y='total_bill', data=df)
sns.set_style("ticks")
sns.set_context("paper")
```

```python
sns.lineplot(x = 'size', y='total_bill', data=df)
sns.set_style("ticks")
sns.set_context("notebook")
```

```python
sns.lineplot(x = 'size', y='total_bill', data=df)
sns.set_style("ticks")
sns.set_context("talk")
```

```python
sns.lineplot(x = 'size', y='total_bill', data=df)
sns.set_style("ticks")
sns.set_context("poster")
```

```python
sns.lineplot(x = 'size', y='total_bill', data=df)
sns.set_style("ticks")
sns.set_context("poster")
sns.despine()
```

### Distplot

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['total_bill'], kde=False)
```

```python
sns.distplot(df['total_bill'], hist=False)
```

### Scatter Plot

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
sns.set_palette('vlag')
```

### Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex', kind='scatter', palette='husl')
sns.set_style('ticks')
```

### Bar Plot

```python
sns.barplot(x='sex', y='total_bill', data=df)
```

```python
sns.barplot(x='sex', y='tip', data=df)
```

### Count Plot

```python
sns.countplot(x='sex', data=df)
```
