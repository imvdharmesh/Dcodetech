```python
import cv2
import os
```

```python
video = cv2.VideoCapture('dataset/Bunny.mp4')

if (video.isOpened() == False):
    print('Error in Openning the Video')

while(video.isOpened()):
    res, frame = video.read()
    
    if res == True:
        cv2.imshow('Frame', frame)
        
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break
        
video.release()
cv2.destroyAllWindows()
```

## Capturing Frame from a Video

```python
video = cv2.VideoCapture('dataset/Bunny.mp4')

current_frame = 0

while True:
    res, frame = video.read()
    
    if res:
        name = 'frame'+str(current_frame)+'.jpg'
        print('Creating new frame...',name)
        cv2.imwrite(os.path.join('frames/',name), frame)
        current_frame += 1
    else:
        break
video.release()
cv2.destroyAllWindows()
```
