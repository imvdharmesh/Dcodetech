# Introduction To Computer Vision

Computer Vision is a discipline that studies how to reconstruct, interrupt and understand a 3D scene from the images.

Computer vision is concerned with modelling and replicating human vision using computer software and hardware

## Image Processing VS Computer Vision

### Image Processing

1) Image processing studies image to image transformation, It is done by enhancing the image by tuning many parameters and features of the images.

2) Here the transformation is applied to an input image and the result image is returned as an output.

3) Image processing is mainly focused on processing the images, enhance them and prepare them to do other tasks.

4) Image processing is a subset of computer vision.

### Computer Vision

1) Computer Vision is focused on extracting information from the input image or videos to have a proper understanding of them to predict the visual input like a human brain.

2) Image Processing is one of the methods that is used for compute vision along with Machine learning techniques.

3) Computer Vision is a super set of Image Processing

4) Example are: Face detection, Hand-writing detection, Object detection

### Image Functions of OpenVC

__imread() :__ This function is for reading an image. it supports various images format such as PNG, JPEG, JPG, TIFF etc.

__imshow() :__ This function is for showing an image in the window. window automatically fits to the image size.

__imwrite() :__ This function is used for writing an image, supports png,jpeg,tiff,etc.

## Text in Image

```python
img = cv2.imread('dataset/lion.jpg')
font = cv2.FONT_HERSHEY_SIMPLEX
pos = (50,50)
fontscale = 1
color = (255,0,0)
thickness = 3
# image = cv2.putText(img, 'Lion', pos, font, fontscale, color, thickness, cv2.LINE_AA, False)
image = cv2.putText(img, 'Lion', pos, font, fontscale, color, thickness, cv2.LINE_AA, True)
cv2.imshow('test', image)
cv2.waitKey(15000)
cv2.destroyAllWindows()
```

## Line in Image

```python
img = cv2.imread('dataset/flower.jpeg')
start_point = (0,0)
end_point = (700,700)
color = (0,0,0)
thickness = 3
image = cv2.line(img, start_point, end_point, color, thickness)
cv2.imshow('Line', image)
cv2.waitKey(15000)
cv2.destroyAllWindows()
```

## Blurr Image

```python
img = cv2.imread('dataset/flower.jpeg')
blur = cv2.GaussianBlur(img, (7,7), 0)
cv2.imshow('Blur', blur)
cv2.waitKey(15000)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('dataset/flower.jpeg')
blur2 = cv2.medianBlur(img, 5)
cv2.imshow('Blur', blur2)
cv2.waitKey(15000)
cv2.destroyAllWindows()
```

## Rotate Image

```python
img = cv2.imread('dataset/flower.jpeg')
(rows, cols) = image.shape[:2]
matrix = cv2.getRotationMatrix2D((cols/2,rows/2), 45, 1)
result = cv2.warpAffine(img, matrix, (cols, rows))
cv2.imshow('Rotation', result)
cv2.waitKey(15000)
cv2.destroyAllWindows()
```
