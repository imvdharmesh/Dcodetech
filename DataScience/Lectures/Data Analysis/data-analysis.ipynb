## Covid Cases Rates

```python
total_cases = covid_df['new_cases'].sum()
```

```python
total_deaths = covid_df['new_deaths'].sum()
```

```python
total_test = covid_df['new_tests'].sum()
```

```python
print(f'Total new cases are : {total_cases}')
print(f'Total new deaths are : {total_deaths}')
print(f'Total new tests are : {total_test}')
```

```python
death_rate = (total_deaths / total_cases) * 100
```

```python
print('Death rate of Italy is {:.2f} %'.format(death_rate))
```

A total of 935310 test were conducted before daily test reports were reported.

```python
initial_test = 935310
total_test_done = total_test + initial_test
total_test_done
```

```python
total_cases
```

```python
positivity_rate = (total_cases / total_test_done) * 100
```

```python
print('Positivity rate is {:.2f} %'.format(positivity_rate))
```
