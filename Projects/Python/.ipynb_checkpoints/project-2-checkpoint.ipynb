## Project - 02

**Mini Inventory Management System.**

    Requirement: You need to create a Mini Inventory Management System using Tkinter and any Database (MySQL,SQLite) preffered.
    
    You should have the following fields in the form and also in your db table

    1) Product id
    2) Product Name
    3) Selling Price
    4) Quantity

    There should also be a view where all the data inserted should be viewd from the database table.

    All sorts of validation can be done in the form, you can use any logic as per your need.

    Also you need to keep a check that the product id should not be duplicate.

    The product name can be same but they should have different product id.

### Database & Table

```python
import sqlite3

con = sqlite3.connect('mims.db')
cur = con.cursor()
cur.execute("""CREATE TABLE IF NOT EXISTS inventory_tbl
            (product_id VARCHAR(128) NOT NULL,
            product_name VARCHAR(200) NOT NULL,
            sell_price DECIMAL(10,2) NOT NULL,
            quantity INT NOT NULL,
            PRIMARY KEY(product_id))""")
print("Table created successfully")
cur.close()
con.close()
```

### Program :

```python
import tkinter as tk
from tkinter import *
from tkinter import ttk,scrolledtext,messagebox
import sqlite3

win = tk.Tk()
win.title('Mini Inventory Management System')
tabcontrol = ttk.Notebook(win)
inventory = ttk.Frame(tabcontrol)
lblframe = ttk.LabelFrame(inventory,text="Inventory Management")
lblframe.grid(row=0,column=0,padx=8,pady=4,sticky="N")


lblframe1 = ttk.LabelFrame(inventory, text='Product List', borderwidth=3)
lblframe1.grid(row=0,column=1,padx=8,pady=4,sticky="N")


tree = ttk.Treeview(lblframe1, columns=('Product ID','Product Name','Selling Price','Quantity'), height=20)
tree.place(x=30,y=95)
vsb = ttk.Scrollbar(tree, orient="vertical",command=tree.yview)
vsb.place(x=30+954+5,y=0,height=200+220)

tree.configure(yscrollcommand=vsb.set)
tree.heading('#0',text='Item No')
tree.heading('#1',text='Product ID')
tree.heading('#2',text='Product Name')
tree.heading('#3',text='Selling Price')
tree.heading('#4',text='Quantity')
tree.column('#0',stretch=tk.YES)
tree.column('#1',stretch=tk.YES)
tree.column('#2',stretch=tk.YES)
tree.column('#3',stretch=tk.YES)
tree.column('#4',stretch=tk.YES)
tree.grid(row=10, columnspan=4,sticky='nsew')


def insert_data():
    try:
        if P_ID.get() !="" and P_NAME.get() !="" and P_PRICE.get()!="" and P_QUANT.get()!="":
            con = sqlite3.connect('mims.db')
            cur = con.cursor()
            cur.execute('INSERT INTO inventory_tbl(product_id,product_name,sell_price,quantity) VALUES(?,?,?,?)',
                        [P_ID.get(),str(P_NAME.get()).lower(),P_PRICE.get(),P_QUANT.get()])
            con.commit()
            messagebox.showinfo("","Record Inserted Successfully.")
            P_ID.set("")
            P_NAME.set("")
            P_PRICE.set("")
            P_QUANT.set("")

        else:
            messagebox.showinfo("","Please fill all fields.")
    
    except tk.TclError:
        messagebox.showinfo("","Price & Quantity should be a number.")
        
    except sqlite3.IntegrityError:
        messagebox.showwarning("","Product ID already exist.")

def show_data():
    try:
        i=0
        tree.delete(*tree.get_children())
        con = sqlite3.connect('mims.db')
        cur = con.cursor()
        record = cur.execute('SELECT * FROM inventory_tbl')
        for row in record:
            tree.insert('','end',text="Item "+str(i),values=(row[0],row[1],row[2],row[3]))
            i=i+1
        cur.close()
    except:
        messagebox.showwarning("","No Data Found.")

global P_ID
global P_NAME
global P_PRICE
global P_QUANT

P_ID = tk.StringVar()
P_NAME = tk.StringVar()
P_PRICE = tk.DoubleVar()
P_PRICE.set("")
P_QUANT = tk.IntVar()
P_QUANT.set("")

pro_id = ttk.Label(lblframe,text="Product ID:")
pro_id_tb = ttk.Entry(lblframe,textvariable=P_ID,width=25)
pro_name = ttk.Label(lblframe,text="Product Name:")
pro_name_tb = ttk.Entry(lblframe,textvariable=P_NAME,width=25)
pro_price = ttk.Label(lblframe,text="Selling Price:")
pro_price_tb = ttk.Entry(lblframe,textvariable=P_PRICE,width=25)
pro_quant = ttk.Label(lblframe,text="Quantity:")
pro_quant_tb = ttk.Entry(lblframe,textvariable=P_QUANT,width=25)
insertButton = ttk.Button(lblframe,text="Insert",command=insert_data)
showButton = ttk.Button(lblframe,text="Show",command=show_data)

style = ttk.Style()
style.configure('TButton',background='#3498db')

pro_id.grid(column=0,row=0,sticky='W')
pro_id_tb.grid(column=0,row=1,sticky='W')
pro_name.grid(column=0,row=2,sticky='W')
pro_name_tb.grid(column=0,row=3,sticky='W')
pro_price.grid(column=0,row=4,sticky='W')
pro_price_tb.grid(column=0,row=5,sticky='W')
pro_quant.grid(column=0,row=6,sticky='W')
pro_quant_tb.grid(column=0,row=7,sticky='W')
insertButton.grid(column=0,row=8,sticky='W',pady=7)
showButton.grid(column=0,row=8,sticky='E')

tabcontrol.pack(expand=1,fill="both")
inventory.pack()

win.mainloop()
```
