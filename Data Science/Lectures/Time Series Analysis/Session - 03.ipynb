# DataFrame in Time Series Analysis

```python
import numpy as np
import pandas as pd
```

```python
frame = pd.DataFrame(np.random.randn(24,4),
                     index=pd.period_range('1-2020','12-2021',freq='M'),
                    columns=['Mumbai','Pune','Chennai','Hydrabad'])
```

```python
frame
```

```python
dec_sample = frame.resample('A-DEC')
```

```python
dec_sample.min()
```

```python
dec_sample.max()
```

```python
dec_sample.mean()
```

```python
dec_sample.sum()
```

```python
dec_sample.count()
```

```python
dec_sample.median()
```

```python
quarter = frame.resample('Q-DEC')
```

```python
quarter.min()
```

```python
quarter.max()
```

```python
quarter.mean()
```

```python
quarter.sum()
```

## Filling Null Values

```python
df = pd.DataFrame({'A':[10,20,None,30],
                 'B':[None,2,4,6],
                 'C':[4,12,16,5],
                 'D':[5,4,7,None]})
```

```python
df
```

```python
df.ffill()  # column wise
```

```python
df.ffill(axis=1)  # row wise
```

```python
res = df.ffill()
```

```python
res
```

```python
res.ffill(axis=1)
```

```python
df.bfill()
```

```python
df.bfill(axis=1)
```

```python
df.fillna(method='bfill')
```
